```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
"GridSearchCV me hum ye check karte hain ki kis hyperparameter par best result aayega."
```

```python
df = sns.load_dataset('iris')#preinstalled
df.head()#species ke alwa baki sab pahle se hi numeric me hai
```

```python
df['species'].unique()# kaun kau nspecies hai and ise predict karna hai
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('species',axis = 1)#species ko drop kar rhe hai kyuki usi ko predict karna hai isliye baki sbaki baad me rakhe hai
y = df['species']#axis =1 ka mtlb hota hai column operation by default axis=0 hota hai yani ki row operation
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
# bas xtest likho and summary se nikal lo sabme use kar sakte hai
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(n_neighbors=15)
model_knn.fit(X_train,y_train)
#agar n=5 diye tu accuaracy hai 98% and wahi agar 15 kiye tu accuracy hai 100% (but ye overfit hai)
#
```

```python
#ye pahle se hi pretrained data hai isliye itna jayda accuracy aa rhi hai
model_knn.score(X_test,y_test)
```

```python
#svm
from sklearn.svm import SVC
# model_svm = SVC(C=30,kernel="rbf",gamma="auto")
model_svm = SVC(gamma="auto")#C ko punishment maan sakte ho
model_svm.fit(X_train,y_train)
#similarly agar isme hm c ki value 30 rakhe tu 98% hai wahi agar 10 karde tu 100 hai
#iska mtlb multiple hyperparametr hota hai and kisi hyperparameter ko use kare tu wo change hota hai.
```

```python
model_svm.score(X_test,y_test)
```

```python

#now lets use the grid search CV
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier = GridSearchCV((model_svm),{#hmne jo model banaya hai svm ka usko daal diye gridsearchcv me
    'C' : [1,10,20,30],#hm jab test kar rhe the tab C and kernel ki value pahle hi de diye the jisse wo answer de rha tha
    'kernel' :['rbf','linear'],# ab hm ise hta ke check karnege
},cv = 5,return_train_score = False)
```

```python
classifier.fit(X,y)# isme khud hi bta rha hai ki best kaunsa hai
```

---

### ❓ Kya Cross Validation me `train_test_split()` hota hai?

❌ **Nahi hota**, kyunki Cross Validation (jaise K-Fold) **khud hi data ko multiple train-test parts me split** karta hai.

➡️ Isliye hume manually `train_test_split()` karne ki zarurat nahi padti.

```python
classifier.cv_results_
```

```python
results = pd.DataFrame(classifier.cv_results_)
results
```

```python
#jisko dekhna hai use paste kar du
results[['param_C','param_kernel','mean_test_score']]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
classifier_r = RandomizedSearchCV((model_svm),{#is model me lekin grid hi sahi hai kyuki ye chota hai
    'C' : [1,10,20,30],
    'kernel' :['rbf','linear'],
},n_iter = 4,return_train_score = False)
```

```python
classifier_r.fit(X,y)
```

```python
results = pd.DataFrame(classifier_r.cv_results_)
results
```
