**Basic project to understand the concept**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df.columns
```

```python
df.info()
#deck column me bahut sare hi null value hai isliye idkodrop kar denge
```

```python
df.drop(["deck", "embark_town", "alive", "class", "who", "adult_male"], axis=1, inplace=True)
df.info()
```

```python
#ages ko fill akr rhe hai
df["age"].fillna(df["age"].mean(), inplace=True)
#null values se
```

```python
df.dropna(subset=["embarked"], inplace=True)
#puri ki puri row ko drop karne ke liye
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
#only sex and embarked is in strings
```

```python
df.head()
```

```python
df['sex'] = le.fit_transform(df['sex'])
df["embarked"] = le.fit_transform(df["embarked"]) # S=2, C=0, Q=1
```

```python
df.head()
```

```python
df = df.astype(int)
```

```python
df.head()
```

```python
x= df.drop("survived", axis=1)
y= df["survived"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
#hmara model acha hai ya nhi ye pta karne ke liye hm use karte hai confusion matrix logistic regression me
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score,classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
#19 and 16 wrong  prediction
```

```python
print(classification_report(y_test,y_pred))#bina print ke kharab dikh rha tha
```

```python
#logistic regression se accuracy 80% aayi hai.
```

**To ise sahi kaarne ke liye hm standard scaler lga ke feature scaling karenge.**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
#hme age fare and pclsss ko sahi karna hai, hm sare value ko bhi standard scale kar sakte hai jo value 0 and 1 hai wo asie hi rah jayenge
#and jo nhi hai wo ban jayenge
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
#fit_transform()  Short Notes
# fit()  Calculate stats (mean, std, min-max, etc.)

# transform()  Use those stats to scale the data

# fit_transform() = fit + transform (used on training data)
# Why Use?
#  Scale all features to same range

#  Avoid model bias due to different value scales

#  Important for KNN, Logistic, SVM, etc.
```

```python
from sklearn.neighbors import KNeighborsClassifier#dont need to remember everuthing learn using documnetation
```

```python
knn_model = KNeighborsClassifier(n_neighbors= 5)
knn_model.fit(x_train_scaled,y_train)#fit train ke liye
```

```python
knn_model.score(x_train_scaled,y_train)# score Checks model accuracy on given data
```

```python
y_pred_knn=knn_model.predict(x_train_scaled)
```

```python
y_pred_knn = knn_model.predict(x_test_scaled)
accuracy_score(y_test,y_pred_knn)
```

```python
#logistic regression ka 80% tha and iska 77% hai.
```

```python
confusion_matrix(y_test,y_pred_knn)
```
